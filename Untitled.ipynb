```python
import numpy as np
import pandas as pd
from tsfresh.feature_extraction import feature_calculators as fc
from pandas.api.indexers import BaseIndexer
```

```python
def time_data(duration):
    acquisition_frequency = 100 #Hz
    time_data = [0.01 + time_step/acquisition_frequency for time_step in range(0,duration)]
    return(time_data)
```

```python
np.random.seed(0)
df = pd.DataFrame({ 'B':list(np.random.randint(low = 0,high=100,size=2000)),
                    'C':list(np.random.randint(low = 0,high=100,size=2000)),
                    'D':list(np.random.randint(low = 0,high=300,size=2000))})
```

```python
df['time'] = time_data(df.shape[0])
```

```python
df['abs_energy'] = fc.abs_energy(np.array(df['B']))
```

```python
df.shape
```

```python
def temp_window(df, size, step):
    
    max_abs = pd.Index(df.index.values).max()
    
    start =[]
    end = []
    
    i = pd.Index(df.index.values).min()
    s = step
    
    start.append(i)
    end.append(size)
    j=1
    
    while size < max_abs:
        i = step
        size = size + s
        step = step + s
        if size > max_abs:
            size = max_abs
        start.append(i)
        end.append(size)
        j += 1
    start = np.array(start)
    end = np.array(end)
    
    return(start, end)
```

```python
def aprox_entropy(df, col, index_windows):
    
    entropy = []
    ind_i, ind_f = index_windows
    
    col = df.columns.get_loc(col)
    
    for i in range(len(ind_i)):
        tw = df.iloc[ind_i[i]:ind_f[i],col]
        entropy.append(fc.approximate_entropy(np.array(tw), m=10,r=2))
    
    data = {}
    data['aprox_entropy'] = entropy
    entropy_df = pd.DataFrame(data)
    return entropy_df

t_ind = temp_window(df,50,10)
entropy_df = aprox_entropy(df,'B', t_ind)
entropy_df.plot()
```
